```python

import pandas as pd
import numpy as np



df = pd.read_csv('E:\Code\PycharmProjects\Projects\二手车实战\data\car.csv',encoding='gbk')
df
```

```python
# 删除具有缺失值的行
df = df.dropna()
df.info()
```

```python
# 统计汽油号数
df['oil_num'].value_counts()
```

```python
origin_price = df.sort_values(by='origin_price')
origin_price
```

```python
# 价格排行前五
df1 = df.sort_values(by='price',ascending=False) # 按照现价排序
df11 = df1.drop_duplicates(['title']) # 对title列去重
df111  =df11[['title','price']][:10]
df111
```

```python
# 二手车数量前十的城市
df2 = df['city'].value_counts()[:10]
print(df2.index.tolist())
print(df2.values.tolist())
# df2
```

```python
# 排行前十的品牌
df3 = df['title'].value_counts()[:10]
print(df3.index.tolist())
print(df3.values.tolist())
```

```python
# 原价排行前十与二手价对比
a = df.sort_values('origin_price',ascending=False).drop_duplicates(['title'])
a1 = a[['title','origin_price','price']].head(10)
a1_title = a1['title'].tolist()
a1_origin = a1['origin_price'].tolist()
a1_now = a1['price'].tolist()

print(a1_title)
print(a1_origin)
print(a1_now)


import matplotlib.pyplot as plt
plt.rcParams['font.family']='simhei'
plt.rcParams['axes.unicode_minus']=False

plt.title('原价排行前十与二手价对比',fontsize=16)
plt.plot(a1_title,a1_origin,label='原价')
plt.plot(a1_title,a1_now,label='二手价')
plt.legend(loc='center left',bbox_to_anchor=[1,0.5,0.5,1])
plt.show()
```
